```python
import numpy as np
import skfuzzy.control as ctrl
temp = np.linspace(20.4, 41.2, num=100)
wind_speed = np.linspace(0.7, 3.1, num=100)
rain_fall = np.linspace(0.2, 9.1, num=100)
temp = ctrl.Antecedent(temp, 'temp')
wind_speed = ctrl.Antecedent(wind_speed, 'wind_speed')
rain_fall = ctrl.Consequent(rain_fall, 'rain_fall')
names_temp = ['Very Low','Low', 'Normal', 'Hight','Very Hight']
names_wind_speed = ['Very Low','Low', 'Normal', 'Hight','Very Hight']
names_rain_fall = ['Very Low','Low', 'Normal', 'Hight','Very Hight']
temp.automf(names=names_temp)
wind_speed.automf(names=names_wind_speed)
rain_fall.automf(names=names_rain_fall)
rule1 = ctrl.Rule(wind_speed['Very Low'] & temp['Very Low'], rain_fall['Very Low'])
rule2 = ctrl.Rule(wind_speed['Low'] & temp['Low'], rain_fall['Low'])
rule3 = ctrl.Rule(wind_speed['Normal'] & temp['Normal'], rain_fall['Normal'])
rule4 = ctrl.Rule(wind_speed['Hight'] & temp['Hight'], rain_fall['Hight'])
rule5 = ctrl.Rule(wind_speed['Very Hight'] & temp['Very Hight'], rain_fall['Very Hight'])
rule6 = ctrl.Rule(wind_speed['Very Low'] & temp['Low'], rain_fall['Low'])
rule7 = ctrl.Rule(wind_speed['Very Low'] & temp['Normal'], rain_fall['Normal'])
rule8 = ctrl.Rule(wind_speed['Very Low'] & temp['Hight'], rain_fall['Normal'])
rule9 = ctrl.Rule(wind_speed['Very Low'] & temp['Very Hight'], rain_fall['Normal'])
rule10 = ctrl.Rule(wind_speed['Low'] & temp['Very Low'], rain_fall['Low'])
rule11 = ctrl.Rule(wind_speed['Low'] & temp['Normal'], rain_fall['Normal'])
rule12 = ctrl.Rule(wind_speed['Low'] &temp['Hight'], rain_fall['Hight'])
rule13 = ctrl.Rule(wind_speed['Low'] & temp['Very Hight'], rain_fall['Hight'])
rule14 = ctrl.Rule(wind_speed['Normal'] & temp['Very Low'], rain_fall['Low'])
rule15 = ctrl.Rule(wind_speed['Normal'] & temp['Low'], rain_fall['Normal'])
rule16 = ctrl.Rule(wind_speed['Normal'] & temp['Hight'], rain_fall['Hight'])
rule17 = ctrl.Rule(wind_speed['Normal'] & temp['Very Hight'], rain_fall['Hight'])
rule18 = ctrl.Rule(wind_speed['Hight'] & temp['Very Low'], rain_fall['Normal'])
rule19 = ctrl.Rule(wind_speed['Hight'] & temp['Low'], rain_fall['Normal'])
rule20 = ctrl.Rule(wind_speed['Hight'] & temp['Normal'], rain_fall['Hight'])
rule21 = ctrl.Rule(wind_speed['Hight'] & temp['Very Hight'], rain_fall['Very Hight'])
rule22 = ctrl.Rule(wind_speed['Very Hight'] & temp['Hight'],rain_fall['Very Hight'])
rule23 = ctrl.Rule(wind_speed['Very Hight'] & temp['Normal'], rain_fall['Hight'])
rule24 = ctrl.Rule(wind_speed['Very Hight'] & temp['Low'], rain_fall['Normal'])
rule25 = ctrl.Rule(wind_speed['Very Hight'] &temp['Very Low'],rain_fall['Normal'])
system_ctrl= ctrl.ControlSystem(rules=[rule1, rule1, rule2, rule3, rule4, rule5,rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25])
system=ctrl.ControlSystemSimulation(system_ctrl)
system.input['temp'] = 22.5
system.input['wind_speed'] = 2.2
system.compute()
print(system.output['rain_fall'])
rain_fall.view(sim=system)

```
