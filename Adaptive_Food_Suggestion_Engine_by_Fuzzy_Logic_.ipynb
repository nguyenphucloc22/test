```python
# Bài 2 Adaptive Food Suggestion Engine by Fuzzy Logic 
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl

#INPUT
x_taste=np.linspace(0,3,num=100)
x_enviroment=np.linspace(0,4,num=100)
x_budget=np.linspace(0,9,num=100)
x_review=np.linspace(0,5,num=100)
x_location=np.linspace(0,4,num=100)
x_preference=np.linspace(0,15,num=100)

taste=ctrl.Antecedent(x_taste,"taste")
enviroment=ctrl.Antecedent(x_enviroment,"enviroment")
budget=ctrl.Antecedent(x_budget,"budget")
review=ctrl.Antecedent(x_review,"review")
location=ctrl.Antecedent(x_location,'location')
preference=ctrl.Consequent(x_preference,"preference")

names_taste=['low','medium','high']
names_enviroment=['low','medium','high']
names_budget=['low','medium','high']
names_review=['very_low',"low",'medium','high','very_high']
names_location=['low','medium','high']
names_preference=['level1','level2','level3','level4','level5']
preference["level1"]=fuzz.trimf(preference.universe,[0,3,3])
preference["level2"]=fuzz.trimf(preference.universe,[3,6,9])
preference["level3"]=fuzz.trimf(preference.universe,[6,9,12])
preference["level4"]=fuzz.trimf(preference.universe,[9,12,15])
preference["level5"]=fuzz.trimf(preference.universe,[12,15,15])

taste.automf(names=names_taste)
enviroment.automf(names=names_enviroment)
budget.automf(names=names_budget)
location.automf(names=names_location)
review.automf(names=names_review)
preference.automf(names=names_preference)

rule1=ctrl.Rule(taste['low'] & enviroment['low']& budget['low']& location['low'] & review['very_low'], preference['level1'])
rule2=ctrl.Rule(taste['low'] & enviroment['medium']& budget['low']& location['low'] & review['very_low'], preference['level1'])
rule3=ctrl.Rule(taste['low'] & enviroment['high']& budget['low']& location['low'] & review['very_low'], preference['level2'])
rule4=ctrl.Rule(taste['low'] & enviroment['low']& budget['medium']& location['low'] & review['low'], preference['level1'])
rule5=ctrl.Rule(taste['low'] & enviroment['low']& budget['high']& location['low'] & review['low'], preference['level1'])
rule6=ctrl.Rule(taste['low'] & enviroment['low']& budget['low']& location['medium'] & review['low'], preference['level1'])
rule7=ctrl.Rule(taste['low'] & enviroment['low']& budget['low']& location['high'] & review['low'], preference['level1'])
rule8=ctrl.Rule(taste['low'] & enviroment['high']& budget['low']& location['low'] & review['medium'], preference['level3'])
rule9=ctrl.Rule(taste['low'] & enviroment['low']& budget['low']& location['low'] & review['high'], preference['level2'])
rule9=ctrl.Rule(taste['low'] & enviroment['medium']& budget['low']& location['low'] & review['high'], preference['level3'])
rule10=ctrl.Rule(taste['low'] & enviroment['medium']& budget['high']& location['medium'] & review['high'], preference['level3'])
rule11=ctrl.Rule(taste['medium'] & enviroment['low']& budget['low']& location['low'] & review['medium'], preference['level2'])
rule12=ctrl.Rule(taste['medium'] & enviroment['medium']& budget['low']& location['low'] & review['medium'], preference['level3'])
rule13=ctrl.Rule(taste['medium'] & enviroment['high']& budget['low']& location['low'] & review['medium'], preference['level4'])
rule14=ctrl.Rule(taste['medium'] & enviroment['low']& budget['medium']& location['low'] & review['medium'], preference['level3'])
rule15=ctrl.Rule(taste['medium'] & enviroment['low']& budget['high']& location['low'] & review['medium'], preference['level3'])
rule16=ctrl.Rule(taste['medium'] & enviroment['low']& budget['low']& location['medium'] & review['high'], preference['level1'])
rule17=ctrl.Rule(taste['medium'] & enviroment['medium']& budget['medium']& location['high'] & review['high'], preference['level4'])
rule18=ctrl.Rule(taste['medium'] & enviroment['medium']& budget['high']& location['low'] & review['high'], preference['level5'])
rule19=ctrl.Rule(taste['medium'] & enviroment['high']& budget['low']& location['low'] & review['high'], preference['level4'])
rule20=ctrl.Rule(taste['medium'] & enviroment['high']& budget['high']& location['medium'] & review['high'], preference['level5'])
rule21=ctrl.Rule(taste['high'] & enviroment['low']& budget['low']& location['low'] & review['medium'], preference['level3'])
rule22=ctrl.Rule(taste['high'] & enviroment['medium']& budget['low']& location['low'] & review['medium'], preference['level3'])
rule23=ctrl.Rule(taste['high'] & enviroment['high']& budget['low']& location['low'] & review['medium'], preference['level4'])
rule24=ctrl.Rule(taste['high'] & enviroment['low']& budget['medium']& location['low'] & review['high'], preference['level4'])
rule25=ctrl.Rule(taste['high'] & enviroment['high']& budget['high']& location['low'] & review['medium'], preference['level3'])
rule26=ctrl.Rule(taste['high'] & enviroment['low']& budget['low']& location['medium'] & review['high'], preference['level4'])
rule27=ctrl.Rule(taste['high'] & enviroment['medium']& budget['medium']& location['high'] & review['very_high'], preference['level5'])
rule28=ctrl.Rule(taste['high'] & enviroment['medium']& budget['high']& location['low'] & review['very_high'], preference['level5'])
rule29=ctrl.Rule(taste['high'] & enviroment['high']& budget['low']& location['low'] & review['very_high'], preference['level5'])
rule30=ctrl.Rule(taste['high'] & enviroment['high']& budget['high']& location['medium'] & review['very_high'], preference['level5'])
system_ctrl=ctrl.ControlSystem(rules=[rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule30])
system=ctrl.ControlSystemSimulation(system_ctrl)
system.input['taste'] = 3
system.input['enviroment'] = 1
system.input['budget'] = 2
system.input['location'] = 3
system.input['review'] = 3
system.compute()
print(system.output['preference'])
preference.view(sim=system)

```
